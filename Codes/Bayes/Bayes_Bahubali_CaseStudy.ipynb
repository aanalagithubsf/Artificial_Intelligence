H - Hypothesis : Babubali is a great warrior (Existing belief) (Prior Belief) <br/>
E - Event/Evidence : Bahubali winnning the battle against enemy<br/>
P(H) - Probability of Babubali is a great warrior that is already known<br/>
P(E) - Probability of Babubali winning the battle against enemy<br/>
P(H|E) - Probalility of Babubali is a great warrior given that Babubali wins the battle. This means the new probability of Bahubali being a great warrior after observing an event or evidence. (New Belief) (Posterior Belief)<br/>
P(E|H) - Probability of Babubali winning the battle against enemy given that Bahubali is a great warrior<br/><br/><br/>


P(E) = P(H) P(E|H) + P(~H) P(E|~H)<br/>
P(E) is the complete Probability of Babubali winning the battle against enemy, no matter whether Bahubali is great warrior or not. It might be possible that Bahubali wins the battle by luck.<br/>

```python
import pandas as pd
import numpy as np
```

# Battle with Bandits

```python
P_H = 0.001 # Begins with assumption
P_Not_H = 1 - P_H
P_E_given_H = 0.90 # assumption
P_E_given_Not_H = 1 - P_E_given_H
```

```python
P_E = P_H * P_E_given_H + P_Not_H * P_E_given_Not_H
print(P_E)
```

```python
P_H_given_E = (P_E_given_H * P_H) / P_E
print('Old Belief that Babubali is a great warrior ', P_H )
print('New Belief that Babubali is a great warrior after winning the battle ', P_H_given_E )
```
